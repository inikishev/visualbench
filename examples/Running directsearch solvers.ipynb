# directsearch
We will run directsearch on an image registration problem.

````python
import visualbench as vb
import directsearch
from IPython.display import Video
````

We can get initial parameters of a benchmark concatenated to a vector like this:

```
x0 = bench.get_x0().numpy(force=True)
```

To evaluate benchmark at parameters given in vector x, use `fx = bench.loss_at(x)`, `fx` will be a float. This is exactly what directsearch expects, so we can just pass it the `bench.loss_at` method to minimize.

````python
bench = vb.AffineRegistration(vb.data.SANIC96)

print(f"{bench.ndim = }") # number of variables

res = directsearch.solve(
    f = bench.loss_at,
    x0 = bench.get_x0().numpy(force=True),
    maxevals = 1000,
)

print(res)

bench.render("media/directsearch.mp4", scale=2)
Video("media/directsearch.mp4", embed=True, width=1080)
````
