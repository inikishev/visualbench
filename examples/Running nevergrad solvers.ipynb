# Nevergrad
We will run nevergrad's NGOpt optimizer on a gaussian processes problem.

````python
import visualbench as vb
import nevergrad as ng
from IPython.display import Video
````

We can get initial parameters of a benchmark concatenated to a vector like this:

```
x0 = bench.get_x0().numpy(force=True)
```

To evaluate benchmark at parameters given in vector x, use `fx = bench.loss_at(x)`, `fx` will be a float. Since nevergrad provides derivative-free algorithms, this is all we need - we can just pass it the `bench.loss_at` method to minimize.

````python
bench = vb.GaussianProcesses('spiral', 200, resolution=256).cuda()

print(f"{bench.ndim = }") # number of variables

parametrization=ng.p.Array(init=bench.get_x0().numpy(force=True))
optimizer = ng.optimizers.NGOpt(parametrization=parametrization, budget=1000)
solution = optimizer.minimize(bench.loss_at)

bench.render("media/Nevergrad NGOpt.mp4")
Video("media/Nevergrad NGOpt.mp4", embed=True, width=1000)
````
